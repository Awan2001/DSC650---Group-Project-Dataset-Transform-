## Drop unnecessary columns

```python
# Remove columns that are useless for analysis

# Columns from df_restos
print("First df : Restos")
columns_to_drop_restos = ['Location']
for column in columns_to_drop_restos:
    if column in df_restos.columns:
        df_restos = df_restos.drop(columns=[column])
        print(f"Reason for removal :\n- '{column}' : Contains 100% null values.")
    else:
        print(f"You are trying to remove non-existent column : '{column}'")
display(df_restos.head(10))
print("=" * 100 + "\n\n")

# Columns from df_reviews
print("Second df : Reviews")
columns_to_drop_reviews = ['uuid', 'createdAt', 'updatedAt', 'replies', 'overall']
reasons_for_removal = {
    'uuid': "Unique ID (PK) for reference already exists which is the 'StoreId' in both df.",
    'createdAt': "Queries using these columns are less meaningful.",
    'updatedAt': "Queries using these columns are less meaningful.",
    'replies': "Values are too complex and is a short paragraph.",
    'overall': "Values are duplicates from column 'restaurant_food'. 'rider' is an independent column for rating of the delivery person hence not associated with the 'overall' rating as it should."
}
removal_reasons = []
nonexistent_columns = []
for column in columns_to_drop_reviews:
    if column in df_reviews.columns:
        df_reviews = df_reviews.drop(columns=[column])
        removal_reasons.append(f"- '{column}' : {reasons_for_removal[column]}")
    else:
        nonexistent_columns.append(f"'{column}'")
if nonexistent_columns:
    print(f"You are trying to remove non-existent column : {', '.join(nonexistent_columns)}")
if removal_reasons:
    print("Reason for removal :")
    print("\n".join(removal_reasons))
display(df_reviews.head(10))
print("=" * 100)
```

### Due to the text column contain commas and other unique symbol make this hard to analysis. Thefore, it is better to drop this column.

```python
# prompt: drop text column from df_reviews

# Drop the 'text' column from df_reviews
df_reviews = df_reviews.drop(columns=['text'])

# Display the first 10 rows of the updated df_reviews
display(df_reviews.head(10))
```

## Restructure datasets (rename, reformat and rearrange)

```python
# Restructure columns for analysis

# Columns from df_restos
df_restos['Reviewers'] = df_restos['Reviewers'].str.replace(r'[-()+\s]+', '', regex=True)

# Columns from df_reviews
# Rename columns
df_reviews = df_reviews.rename(columns={
    'likeCount': 'Likes',
    'restaurant_food': 'FoodRating',
    'rider': 'RiderRating'
})

# Change dtype of certain columns
df_reviews['FoodRating'] = df_reviews['FoodRating'].astype('float64')
df_restos['Reviewers'] = df_restos['Reviewers'].astype('int64')

# Rearrange columns
df_reviews = df_reviews[['StoreId', 'isAnonymous', 'reviewerId', 'isLiked', 'Likes', 'FoodRating', 'RiderRating']]

# Display both df info
print("First df : Restos")
print("-" * 50)
df_restos.info()
print("=" * 50 + "\n\n")

print("Second df : Reviews")
print("-" * 50)
df_reviews.info()
print("=" * 50)
```

## Review at duplicate records for confirmation

```python
try:
    from tabulate import tabulate
except ImportError:
    print("Installing tabulate...")
    !pip install tabulate
    from tabulate import tabulate

# Print information about duplicate records in df_reviews
print("Duplicate records in df_reviews : " + "\n")

# Find all duplicated rows (excluding the first occurrence) based on all columns
duplicates = df_reviews[df_reviews.duplicated(keep=False)]

# Group duplicates by their unique combinations
duplicate_groups = duplicates.groupby(list(duplicates.columns))

# Get the first 5 unique duplicate groups
unique_duplicates = list(duplicate_groups)[:5]

# Iterate over each unique duplicate group
for idx, (group_name, group_df) in enumerate(unique_duplicates):
    print(f"Unique Duplicate Group {idx + 1} :")
    print("-" * 30)

    # Find all indexes with this duplicate group
    duplicate_indexes = group_df.index.tolist()

    # Total count of similar records
    count_duplicates = len(duplicate_indexes)

    # Print details for Unique Duplicate Group
    print(f"Indexes of duplicates : {duplicate_indexes[:10]}")
    print(f"Total similar records found : {count_duplicates}")

    # Print the duplicate record values
    for index in duplicate_indexes:
        print(tabulate(df_reviews.loc[[index]], headers='keys', tablefmt='pretty'))

    # Separate each unique duplicate group with "=" * 110
    if idx < len(unique_duplicates):
        print("=" * 110)
    print("\n")

# Define ANSI escape sequences for red color and reset
red_alert = '\033[91m'  # Red color
reset = '\033[0m'       # Reset to default

# Print red alert message with circled number
print(f"{red_alert}!!! ALERT !!!{reset}")

# List back total count of unique duplicate groups
total_unique_duplicates = len(duplicate_groups)
print(f"Total count of unique duplicate groups : {total_unique_duplicates}")
```

## Remove null and duplicate records

```python
# Remove rows with null values and duplicates for df_restos
df_restos.dropna(inplace=True)
df_restos.drop_duplicates(inplace=True)
df_restos.reset_index(drop=True, inplace=True)

# Remove rows with null values and duplicates for df_reviews
df_reviews.dropna(inplace=True)
df_reviews.drop_duplicates(inplace=True)
df_reviews.reset_index(drop=True, inplace=True)

# For df_restos
print("First df : Restos")
print("-" * 50)

# Print the updated count of null values in each column
print("Updated count of null values in each column :")
print(df_restos.isnull().sum())
print("*" * 50)

# Print the updated count of duplicate rows
print("Updated count of duplicate rows :")
print(df_restos.duplicated().sum())
print("=" * 50 + "\n\n")

# For df_reviews
print("Second df : Reviews")
print("-" * 50)

# Print the updated count of null values in each column
print("Updated count of null values in each column :")
print(df_reviews.isnull().sum())
print("*" * 50)

# Print the updated count of duplicate rows
print("Updated count of duplicate rows :")
print(df_restos.duplicated().sum())
print("=" * 50)
```

```python
# Get the set of column names for each DataFrame
columns_restos = set(df_restos.columns)
columns_reviews = set(df_reviews.columns)

# Find common columns
common_columns = columns_restos.intersection(columns_reviews)

# Print common columns in a single line
print(f"DataFrames can be integrated or joined using columns : {', '.join(sorted(common_columns))}")
print("*" * 100)

# Define function to map pandas data types to Hive data types
def map_to_hive_type(pandas_type):
    if pandas_type == 'int64' or pandas_type == 'int32':
        return 'INT'
    elif pandas_type == 'float64' or pandas_type == 'float32':
        return 'DOUBLE'
    elif pandas_type == 'object':
        return 'STRING'
    elif pandas_type == 'bool':
        return 'BOOLEAN'
    elif pandas_type == 'datetime64[ns]':
        return 'TIMESTAMP'
    elif pandas_type == 'datetime.date':
        return 'DATE'
    else:
        return 'UNIDENTIFIED'

# For df_restos
print("\nFor df_restos :")
print("-" *100)

# Print headers
print("\nHeaders :")
print(df_restos.columns.tolist())

# Print data types
print("\nData types :")
for col in df_restos.columns:
    hive_type = map_to_hive_type(df_restos[col].dtype.name)
    print(f"{col} : {hive_type}")

# For df_reviews
print("\nFor df_reviews :")
print("-" *100)

# Print headers
print("\nHeaders :")
print(df_reviews.columns.tolist())

# Print data types
print("\nData types :")
for col in df_reviews.columns:
    hive_type = map_to_hive_type(df_reviews[col].dtype.name)
    print(f"{col} : {hive_type}")

print("=" * 100)
```
